```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv('/home/nbuser/library/StudentsPerformance.csv')
data.head
```

```python
plt.figure(figsize=(8,5))
sns.distplot(data['math score'] ,kde=False ,color='red' , bins=30)
plt.ylabel('Frequency')
plt.xlabel('Math Score Distribution')
plt.show()
```

```python
# most students have their math score in the range of 60 to 80.
```

```python
plt.figure(figsize=(8,5))
sns.distplot(data['reading score'] ,kde=False ,color='blue' , bins=30)
plt.ylabel('Frequency')
plt.xlabel('Reading Score Distribution')
plt.show()
```

```python
# most students have their reading score in the range of 60 to 80.
```

```python
plt.figure(figsize=(8,5))
sns.distplot(data['writing score'] ,kde=False ,color='yellow' , bins=30)
plt.ylabel('Frequency')
plt.xlabel('Writing Score Distribution')
plt.show()
```

```python
# most students have their writing score in the range of 60 to 80.
```

```python
# BAR PLOT OF SCORES  ON THE BASIS OF GENDER
plt.figure(figsize=(10,4))
plt.subplot(1,3,1)
sns.barplot(x='gender' ,y='reading score' , data=data)

plt.subplot(1,3,2)
sns.barplot(x='gender' ,y='writing score' , data=data)

plt.subplot(1,3,3)
sns.barplot(x='gender' ,y='math score' , data=data)

plt.show()
```

```python
# males have higher math score than females, whereas females have higher scores in reading and writing than males.
```

```python
# BAR PLOT OF SCORES ON THE BASIS OF RACE/ETHNICITY
plt.figure(figsize=(14,4))
plt.subplot(1,3,1)
sns.barplot(x='race/ethnicity' ,y='reading score' , data=data)
plt.xticks(rotation=90)

plt.subplot(1,3,2)
sns.barplot(x='race/ethnicity' ,y='writing score' , data=data)
plt.xticks(rotation=90)


plt.subplot(1,3,3)
sns.barplot(x='race/ethnicity' ,y='math score' , data=data)
plt.xticks(rotation=90)

plt.show()
```
